```python
import numpy as np
import pandas as pd
```

```python
data = {
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]
  }

```

Create a DataFrame with the following data:  

```python
df = pd.DataFrame(data)
```

Print the first five rows of the DataFrame. 

```python
df.head()
```

Get the summary statistics of the 'Age' and 'Salary' columns.

```python
summary = df[['Age' , 'Salary']]
summary
```

Calculate the average salary of employees in the 'HR' department.

```python
avg_salary = df[df['Department'] == 'HR']['Salary'].mean()

print("Average Salary of HR Department:", avg_salary)
```

Add a new column, 'Bonus', which is 10% of the salary.  

```python
df['Bonus'] = df['Salary'] * 0.10

print(df)
```

Filter the DataFrame to show employees aged between 25 and 30.  

```python
Age_btw =df[(df['Age']>=25)&(df['Age']<=30)]
Age_btw
```

Group the data by 'Department' and calculate the average salary for each department.  

```python
avg_salary_by_dept = df.groupby('Department')['Salary'].mean()

print(avg_salary_by_dept)
```

Sort the DataFrame by 'Salary' in ascending order and save the result to a new CSV file. 

```python
sorted_df = df.sort_values(by='Salary', ascending=True)
```

```python
print(sorted_df)
```

```python
sorted_df.to_csv('sorted_employees.csv', index=False)
```
