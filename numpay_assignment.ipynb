```python
import numpy as np
```

Create a 1D NumPy array with integers from 1 to 20

```python
arr = np.arange(1,21)
print(arr)
```

Calculate the sum, mean, median, and standard deviation of the elements in the array.  


```python
sum_arr = np.sum(arr)
print(sum_arr)
```

```python
mean_arr = np.mean(arr)
print(mean_arr)
```

```python
median_arr = np.median(arr)
print(median_arr)
```

```python
std_dev = np.std(arr)
print(std_dev)
```

Find the indices of elements greater than 10 in the array.

```python
indices = np.where(arr > 10)
print(indices)
```

Create a 2D NumPy array of shape 4 X 4 with numbers ranging from 1 to 16.  

```python
arr2 = np.arange(1,17).reshape(4,4)
print(arr2)
```

Find the transpose of the array. 

```python
transpose_arr = np.transpose(arr2)
print(transpose_arr)
```

Calculate the row-wise and column-wise sums of the array. 

```python
row_sum = np.sum(arr2,axis=0)
print(row_sum)
```

```python
col_sum = np.sum(arr2,axis=1)
print(col_sum)
```

Create two 3 X 3 arrays filled with random integers between 1 and 20.

```python
arr3_1 = np.random.randint(1, 21, size=(3, 3))
print (arr3_1)
```

```python
arr3_2 = np.random.randint(1, 21, size=(3, 3))
print (arr3_2)
```

Perform element-wise addition, subtraction, and multiplication. 

```python
addition = arr3_1 + arr3_2
print (addition)
```

```python
substraction = arr3_1 - arr3_2
print (substraction)
```

```python
multiplication = arr3_1 * arr3_2
print (multiplication)
```

Compute the dot product of the two arrays.

```python
dot_product = np.dot(arr3_1,arr3_2)
print(dot_product)
```
